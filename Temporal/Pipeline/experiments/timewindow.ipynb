Finding the right time window for temporal and spatial dataset. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
TEMPORAL = r'/Users/jbalkovec/Desktop/MDR/Temporal/Pipeline/data/master/final_master.csv'
SPATIAL = r'/Users/jbalkovec/Desktop/MDR/Spatial/processed/ground_cleaned_lind_2021.csv'
```

```python
temporal_df = pd.read_csv(TEMPORAL)
spatial_df = pd.read_csv(SPATIAL)

temporal_df["date"] = pd.to_datetime(temporal_df["date"]).dt.tz_localize(None)
spatial_df["date"] = pd.to_datetime(spatial_df["date"]).dt.tz_localize(None).dt.floor("D")
```

```python
# Span
print("Temporal:", temporal_df["date"].min(), "to", temporal_df["date"].max())
print("Spatial:", spatial_df["date"].min(), "to", spatial_df["date"].max())
```

```python
time_df = pd.merge(temporal_df, spatial_df, on='date', how='inner', suffixes=('_temp', '_spat'))
```

```python
print("Merged date range:", time_df["date"].min(), "to", time_df["date"].max())
print("Total overlapping rows:", len(time_df))
```
