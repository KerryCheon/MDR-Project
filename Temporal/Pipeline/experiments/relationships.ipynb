# Correlation

This notebook will aid us in feature reduction

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
CSV_PATH = "/Users/jbalkovec/Desktop/MDR/Temporal/Pipeline/data/master/final_master.csv"
```

```python
df = pd.read_csv(CSV_PATH)
df.drop(columns=["station_id", "crx_vn", "longitude", "latitude", "SM_label"], inplace=True, errors='ignore')
```

## Standard (Linear) Correlation Heatmap

```python
corr_1 = df.corr(numeric_only=True)

plt.figure(figsize=(12, 10))
sns.heatmap(
    corr_1,
    annot=False,
    fmt=".2f",
    cmap="coolwarm",
    square=True,
    cbar_kws={"shrink": 0.8}
)
plt.title("Correlation Heatmap (Linear)", fontsize=14, pad=15)
plt.tight_layout()
plt.show()
```

```python
corr_2 = df.corr(method='spearman', numeric_only=True)

plt.figure(figsize=(12, 10))
sns.heatmap(
    corr_2,
    annot=False,
    fmt=".2f",
    cmap="coolwarm",
    square=True,
    cbar_kws={"shrink": 0.8}
)
plt.title("Correlation Heatmap (Spearman)", fontsize=14, pad=15)
plt.tight_layout()
plt.show()
```

```python
corr_3 = df.corr(method='kendall', numeric_only=True)

plt.figure(figsize=(12, 10))
sns.heatmap(
    corr_3,
    annot=False,
    fmt=".2f",
    cmap="coolwarm",
    square=True,
    cbar_kws={"shrink": 0.8}
)
plt.title("Correlation Heatmap (Kendall)", fontsize=14, pad=15)
plt.tight_layout()
plt.show()
```

## Plotly

```python
import plotly.express as px

corr_4 = df.corr(numeric_only=True)

fig = px.imshow(
    corr_4,
    text_auto=".2f",
    color_continuous_scale="RdBu_r",
    title="Interactive Correlation Heatmap"
)
fig.show()
```

## Cluster Map

```python
corr_5 = df.corr(numeric_only=True)

sns.clustermap(
    corr_5,
    annot=False,
    fmt=".2f",
    cmap="vlag",
    figsize=(12, 10)
)
plt.suptitle("Clustered Correlation Heatmap", fontsize=14, y=1.02)
plt.show()
```
